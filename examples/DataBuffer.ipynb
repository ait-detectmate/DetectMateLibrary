```python
import pandas as pd
import sys, os
sys.path.append(os.path.join(os.path.dirname(os.getcwd()), "src"))

from components.utils.DataBuffer import DataBuffer
```

```python
data = [ "A", "B", "C", "D", "E", "F"]
```

### One-by-one:

```python
buffer = DataBuffer(mode="no_buf", size=None, process_function=lambda x: x)

for data_point in data:
    print("Processing data:", buffer.add(data_point))
```

### All at once:

```python
buffer = DataBuffer(mode="no_buf", size=None, process_function=lambda x: x)

print("Processing data:", buffer.add(data))
```

### Batch-wise:

```python
buffer = DataBuffer(mode="batch", size=2, process_function=lambda x: x)

for data_point in data:
    print("Processing data:", buffer.add(data_point))
```

### Window-wise:

```python
buffer = DataBuffer(mode="window", size=2, process_function=lambda x: x)

for data_point in data:
    print("Processing data:", buffer.add(data_point))
```
