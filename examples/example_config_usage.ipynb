```python
from components.common.detector import CoreDetectorConfig, DetectorInstance, DetectorVariable
from pydantic import BaseModel

class ExampleDetectorConfig(BaseModel):
    threshold: float = 0.5

conf = CoreDetectorConfig()
conf.add_instance(
    pos="instance_1",
    event=1,
    template="Error at line {line_number}",
    variables=[
        DetectorVariable(pos=1, params=ExampleDetectorConfig(threshold=0.7)),
        DetectorVariable(pos=2, params=ExampleDetectorConfig(threshold=0.3)),
        DetectorVariable(pos="level", params=ExampleDetectorConfig(threshold=0.9))
    ],
)
conf.model_dump()
```
